Both the cut and qcut functions convert columns with continuous values to categorical columns, but they apply different techniques. The cut function divides the entire value range into intervals of the same size, called bins.

The qcut function, the focus is on the number of values that fall into the bins. The qcut function groups the values into bins so that there is approximately the same number of values in each bin. The upper and lower bounds are adjusted accordingly.

```python
# create a Series with values between 0 and 10

A = pd.Series([5,0,2,8,4,10,7])

# cut function

A_binned = pd.cut(A, bins=4)

A_binned
```

```python
A_binned = pd.cut(A, bins=[-1, 3, 6, 10], labels=["small","medium","large"])

A_binned
```

```python
# qcut function

A =pd.Series([1, 4, 2, 10, 5, 6, 8, 7, 5, 3, 5, 9])

A_binned = pd.qcut(A, q=3)

print(A_binned.value_counts())
```

Both the cut and qcut functions convert columns with continuous values to categorical columns, but they apply different techniques. The cut function divides the entire value range into intervals of the same size, called bins.

The qcut function, the focus is on the number of values that fall into the bins. The qcut function groups the values into bins so that there is approximately the same number of values in each bin. The upper and lower bounds are adjusted accordingly.

```python
# create a Series with values between 0 and 10

A = pd.Series([5,0,2,8,4,10,7])

# cut function

A_binned = pd.cut(A, bins=4)

A_binned
```

```python
A_binned = pd.cut(A, bins=[-1, 3, 6, 10], labels=["small","medium","large"])

A_binned
```

```python
# qcut function

A =pd.Series([1, 4, 2, 10, 5, 6, 8, 7, 5, 3, 5, 9])

A_binned = pd.qcut(A, q=3)

print(A_binned.value_counts())
```

```python
# passing a list to the q parameter

A = pd.Series([1, 4, 2, 3, 10, 5, 6, 8, 7, 5, 9, 14])

A_binned = pd.qcut(A, q=[0, 0.50, 0.75, 1])

print(A_binned.value_counts())
```

```python
# passing a list to the q parameter

A = pd.Series([1, 4, 2, 3, 10, 5, 6, 8, 7, 5, 9, 14])

A_binned = pd.qcut(A, q=[0, 0.50, 0.75, 1])

print(A_binned.value_counts())
```

With the cut function, the focus is on the upper and lower bounds of the bins.


With the qcut function, the focus is the number of values in the bins.
